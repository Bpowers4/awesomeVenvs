```python
import numpy as np
```

```python
mu, sigma = 0, 0.2
my_random = np.random.normal(mu, sigma, 10000)
```

```python
print(my_random)
```

```python
import matplotlib.pyplot as plt 
count, bins, ignored = plt.hist(my_random, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2)),
linewidth=3, color='y')
plt.show()
```
